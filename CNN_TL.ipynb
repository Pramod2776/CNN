```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
plt.style.use("fivethirtyeight")
%load_ext tensorboard
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data(path='mnist.npz')
X_train.shape, y_train.shape, X_test.shape, y_test.shape
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")

len(X_train), len(X_test)
X_train[0:]
y_train[0:]
```

```python
(X_train_full, y_train_full), (X_test, y_test) = tf.keras.datasets.mnist.load_data(path='mnist.npz')
```

```python
X_train_full = X_train_full / 255.0
X_test = X_test / 255.0
```

```python
X_valid, X_train = X_train_full[:5000], X_train_full[5000:]
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

```python
tf.random.set_seed(42)
np.random.seed(42)
```

```python
LAYERS = [tf.keras.layers.Flatten(input_shape = [28,28]),
          tf.keras.layers.Dense(300, kernel_initializer='he_normal'),
          tf.keras.layers.LeakyReLU(),
          tf.keras.layers.Dense(100, kernel_initializer='he_normal'),
          tf.keras.layers.LeakyReLU(),
          tf.keras.layers.Dense(10, activation = 'softmax')]
```

```python
model = tf.keras.models.Sequential(LAYERS)
```

```python
model.compile(loss = "sparse_categorical_crossentropy",
              metrics = ["accuracy"],
              optimizer = tf.keras.optimizers.SGD(learning_rate = 0.001))
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs = 30, validation_data = (X_valid, y_valid), verbose = 2)
```

```python
model.summary()
```

```python
model.save('pretrained_mnist_model.h5')
```

```python
def update_even_odd_labels(labels):
  for idx,label in enumerate(labels):
    labels[idx]=np.where(label%2 == 0, 1, 0)
  return labels
```

```python
y_train_bin, y_test_bin, y_valid_bin = update_even_odd_labels([y_train, y_test, y_valid])
```

```python
tf.random.set_seed(42)
np.random.seed(42)

LAYERS = [tf.keras.layers.Flatten(input_shape = [28,28]),
          tf.keras.layers.Dense(300, kernel_initializer='he_normal'),
          tf.keras.layers.LeakyReLU(),
          tf.keras.layers.Dense(100, kernel_initializer='he_normal'),
          tf.keras.layers.LeakyReLU(),
          tf.keras.layers.Dense(2, activation = 'softmax')]
model_1 = tf.keras.models.Sequential(LAYERS)
```

```python
start = time.time()
history = model_1.fit(X_train, y_train_bin, epochs = 30, validation_data = (X_valid, y_valid_bin), verbose = 2)
end = time.time()
print(f"Runtime of the program is {end - start}")
```

```python
lower_pretrained_layers = pretrained_mnist_model.layers[:-1]

# Create a new Sequential model with an Input layer
new_model = tf.keras.models.Sequential()
new_model.add(tf.keras.layers.Input(shape=(28, 28)))

# Add the pretrained layers one by one
for layer in lower_pretrained_layers:
    new_model.add(layer)

# Add the new Dense layer
new_model.add(tf.keras.layers.Dense(2, activation ="softmax"))

# Check the model summary
new_model.summary()
```

```python
def update_even_odd_labels(labels):
  for idx, label in enumerate(labels):
    labels[idx] = np.where(label % 2 == 0, 1, 0)
  return labels
```

```python
y_train_bin, y_test_bin, y_valid_bin = update_even_odd_labels([y_train, y_test, y_valid])
```

```python
new_model.compile(loss = "sparse_categorical_crossentropy",
              metrics = ["accuracy"],
              optimizer = tf.keras.optimizers.SGD(learning_rate = 0.001))
```

```python
# starting time
start = time.time()
history = new_model.fit(X_train, y_train_bin, epochs=10,
validation_data=(X_valid, y_valid_bin), verbose=2)
#ending time
end = time.time()
print(f"Runtime of the program is {end - start}")
```
